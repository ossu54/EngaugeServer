```python
import json as json
import matplotlib.pyplot as plt

def get_datadict(path):
    with open(path) as f:
        return json.load(f)

data1 = get_datadict("data1.json")
data2 = get_datadict("data2.json")

```

```python
# extract the braking time series, sampled every 200ms
series1 = [point["braking"] for point in data1["brakingPoints"]]
series2 = [point["braking"] for point in data2["brakingPoints"]]
```

```python

plt.figure(figsize=(10, 5))
plt.ylabel("Braking intensity")

plt.subplot(1, 2, 1)
plt.plot(series1)

plt.subplot(1, 2, 2)
plt.plot(series2)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter([i for i in range(len(series1))], series1)
plt.scatter([i for i in range(len(series2))], series2)

plt.ylabel("Braking intensity")
plt.show()
```
