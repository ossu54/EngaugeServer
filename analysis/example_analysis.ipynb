```python
import json as json
import matplotlib.pyplot as plt

def get_datadict(path):
    with open(path) as f:
        return json.load(f)

def get_all(sessionpaths):
    sessions = []
    for path in sessionpaths:
        sessions.append(get_datadict(path))
    return sessions

#sessions = get_all(["data/oskar1.1.json", "data/oskar1.2.json", "data/kaspar1.1.json"])
sessions = get_all(["data/new1.json", "data/new2.json", "data/new3.json"])

# extract a list of braking points for each session
braking_sessions = [session["brakingPoints"] for session in sessions]
s_all = [[point["braking"] for point in points] for points in braking_sessions]

#s_all_cleaned = [[x-3 for x in points] for points in s_all]
#s_all = s_all_cleaned
```

```python
def plot_line(series, color, title, linewidth=3.0):
    plt.figure(figsize=(30,5))
    plt.ylabel("Braking intensity")
    plt.plot(series, linewidth=linewidth, color=color)
    plt.title(title)
    #plt.ylim((0,1000))
    plt.show()

plot_line(s_all[0], 'red', "RUN 1")
plot_line(s_all[1], 'green', "RUN 2")
plot_line(s_all[1], 'blue', "RUN 3")
```

```python
plt.figure(figsize=(30, 5))
plt.ylabel("Braking intensity")
plt.plot(s_all[0], linewidth=2.0, color='red')
plt.plot(s_all[1], linewidth=2.0, color='green')
plt.plot(s_all[2], linewidth=2.0, color='blue')
plt.title("RUNS 1, 2, 3")
#plt.ylim((0,1000))
plt.show()
```

```python
def plot_hist(series, color, title, nrbins=50):
    plt.figure(figsize=(30, 5))
    plt.hist(series, color=color, bins=nrbins)
    plt.title(title)
    #plt.xlim((0,1000))
    #plt.ylim((100,200))
    plt.show()

plot_hist(s_all[0], 'red', "Braking measurements distribution RUN 1")
plot_hist(s_all[1], 'green', "Braking measurements distribution RUN 2")
plot_hist(s_all[2], 'blue', "Braking measurements distribution RUN 3")
```
